# AI-Powered-Drug-Discovery-Bootcamp Challenge Track

Welcome to the challenge track for AI-Powered-Drug-Discovery-Bootcamp. Your task is to design and evaluate novel small molecules that are predicted to inhibit CDK4 while avoiding CDK11 binding. The workflow combines MolMIM generation, cheminformatics filtering, and Boltz-2 affinity prediction so you can iterate from seed molecules to ranked candidates.

## The Challenge

You will use NVIDIA MolMIM NIM to generate candidate molecular structures, then assess those candidates with a combination of fast molecular-property filters and target-aware affinity predictions. Useful scoring signals include:

- **Drug-likeness (QED)**: overall suitability of a molecule as a drug candidate.
- **Synthesizability (Synthetic Accessibility)**: estimated ease of synthesis.
- **Solubility and physicochemical range**: practical property windows for small-molecule development.
- **Toxicity and PAINS alerts**: early filters for obvious liabilities.
- **Tanimoto similarity and novelty**: balance scaffold familiarity with new chemical matter.
- **Boltz-2 affinity predictions**: compare CDK4 potency against CDK11 off-target binding.

By combining these scoring techniques with MolMIM's latent-space representation of candidate molecules, you can use numerical optimization methods such as [CMA-ES](https://en.wikipedia.org/wiki/CMA-ES) to steer molecule generation toward useful multi-objective designs.

## Challenge Objectives

1. **Design a custom scoring oracle** that combines chemical properties, novelty, and target-aware affinity signals.
2. **Generate novel molecules** from known CDK4-relevant seeds or your own starting points.
3. **Optimize properties** using CMA-ES or another search strategy.
4. **Evaluate drug potential** by ranking candidates for CDK4 potency, CDK11 avoidance, and developability.

## Success Criteria

A strong submission should explain why the top-ranked candidates are expected to be CDK4-selective, drug-like, synthesizable, and reasonably novel. Keep track of scoring weights and assumptions so your results are interpretable rather than just numerically high.
