## OpenHackathon: Designing Selective, Drug‑like CDK4/6 Inhibitors

### Why Selectivity Matters
- **Therapeutic goal**: Inhibit dysregulated cell‑cycle progression driven by CDK4/6 while preserving normal cellular functions.
- **Selectivity rationale**: Broad or off‑target CDK inhibition can cause dose‑limiting toxicities. Prioritizing CDK4/6 and avoiding CDK11 helps balance efficacy and safety.
- **When CDK4/6 selectivity is needed**: HR+ / HER2‑ breast cancer and other tumors with aberrant cyclin D–CDK4/6 activity benefit from CDK4/6 inhibitors (often combined with endocrine or targeted therapy).
- **Why avoid CDK11**: CDK11 (kinase variants of CDC2L genes) participates in transcription/splicing and cell survival pathways; off‑target inhibition risks systemic toxicity and confounded pharmacology.

### Structural Context: What We Know About CDKs
- **Kinase fold and ATP site**: CDK4, CDK6, and CDK11 share a conserved bilobal kinase architecture with a hinge region often engaged by heteroaromatic “hinge binders.”
- **Opportunities for selectivity**: Subtle differences near the gatekeeper residue, solvent‑front pocket, and back‑pocket/DFG region can be exploited to favor CDK4/6 over CDK11.
- **Solved structures**: Multiple X‑ray and cryo‑EM structures are available in the Protein Data Bank for CDK4/6 and CDK11 kinase domains (including complexes with small‑molecule inhibitors), enabling structure‑based design.
- **Reference compounds**:
  - Marketed CDK4/6 inhibitors: **Palbociclib**, **Ribociclib**, **Abemaciclib** (breast cancer).
  - Historical CDK chemotypes: roscovitine/purines, aminopyrimidines, and other hinge‑binding scaffolds with diverse selectivity profiles.

### Challenge Overview
Design and submit small molecules that are:
- **Potent for CDK4 and CDK6** (on‑target activity)
- **Avoiders of CDK11** (off‑target minimization)
- **Drug‑like and synthesizable** (ADME‑friendly, tractable chemistry)
- **Safe‑leaning and novel** (avoid obvious liabilities; prefer chemical novelty)

Your workflows may combine ligand‑ and structure‑based methods, ML/AI generation, docking, and medicinal chemistry intuition.

### AI and Computational Approaches
- **NVIDIA MolMIM (Molecular Model‑In‑the‑Middle)**
  - Use MolMIM for multi‑objective de novo design conditioned on potency proxies, drug‑likeness, and synthesizability.
  - Seed from known CDK4/6 scaffolds to retain privileged interactions; guide generation with selectivity constraints that penalize features enriched in CDK11 binders.
  - Employ iterative loops: generate → filter → refine, using QSAR/property predictors and rule‑based filters.

- **NVIDIA DiffDock (Structure‑aware Docking via Diffusion)**
  - Dock generated molecules into CDK4, CDK6, and CDK11 structures to evaluate poses, hinge interactions, and pocket fit.
  - Prioritize compounds with favorable poses for CDK4/6 and disfavored interactions in CDK11.
  - Combine pose quality with physics/ML scoring and medicinal chemistry filters; use cross‑target docking to estimate selectivity risk.

- **Putting it together**
  - Start with diverse, drug‑like scaffolds via MolMIM; enforce property ranges (e.g., MW, logP, HBD/HBA, rotors) and SMARTS alerts.
  - Use DiffDock to triage by pose plausibility across CDK4/6 vs CDK11.
  - Layer predictive models for potency/selectivity, drug‑likeness (e.g., QED), synthetic accessibility, toxicity risk, and novelty vs public corpora.
  - Iterate to balance on‑target potency with off‑target avoidance and developability.

### How Submissions Are Evaluated (High‑Level)
Submissions are scored with a composite, multi‑criterion approach. At a high level, we consider:
- **On‑target potency**: Evidence for strong activity toward CDK4/6.
- **CDK11 avoidance and selectivity**: Reduced likelihood of CDK11 engagement and favorable on‑target/off‑target balance.
- **Drug‑likeness and synthesizability**: Chemistry consistent with oral drug space and practical synthesis.
- **Safety indicators**: Lower predicted risk across common in vitro/tox liability endpoints.
- **Novelty/diversity**: Dissimilarity from large public databases (e.g., ChEMBL) and internal redundancy control.

Notes:
- Exact scoring details, thresholds, and any specific model weightings are intentionally abstracted to preserve the integrity of the challenge.
- Visual summaries and CSV outputs highlight top‑ranked compounds and component scores for feedback.

### Example Known CDK4/6 Drug Features to Consider
- Hinge‑binding heterocycles forming 1–2 hydrogen bonds.
- Substituents accessing selectivity‑relevant subpockets near the gatekeeper/solvent front.
- Balanced lipophilicity and polarity for permeability while limiting promiscuity.

### Submission Requirements
- **Quantity**: About 1,000 designed compounds per team.
- **Format**: CSV with columns `compound_id,smiles`.
- **Conventions**:
  - Provide unique, stable IDs per compound (no spaces; use underscores).
  - Use valid, preferably canonical SMILES; remove duplicates.
  - Optional extra columns are allowed but will be ignored by the core evaluation.

### Practical Tips
- Ensure structural diversity to avoid over‑optimization to a single chemotype.
- Exploit structural differences between CDK4/6 and CDK11 rather than relying only on global physicochemical tuning.
- Track rationale: record why top designs are predicted to be selective and drug‑like.
- Use staged filters: quick property/novelty screens → docking/pose checks → higher‑fidelity scoring.

We’re excited to see creative, well‑reasoned designs that push the frontier of selective CDK4/6 inhibition while minimizing CDK11 engagement. Good luck!
