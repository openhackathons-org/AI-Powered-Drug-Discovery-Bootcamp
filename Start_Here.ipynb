The field of drug discovery is increasingly benefiting from AI, helping researchers build efficient predictive and generative workflows that can reduce the time needed to identify promising molecular candidates. NVIDIA BioNeMo NIMs provide deployable services for molecule generation, biomolecular structure prediction, and affinity estimation.

In this bootcamp, you will run a local BioNeMo workflow that generates candidate small molecules with MolMIM, evaluates CDK4/CDK11 binding behavior with Boltz-2, and ranks compounds with practical cheminformatics filters.

### Takeaways

2. Use MolMIM latent-space generation to propose and refine drug-like small molecules.
3. Use Boltz-2 affinity predictions and cheminformatics scores to rank CDK4-selective candidates.
