Welcome to an exciting journey into AI-powered drug discovery! This bootcamp combines comprehensive tutorials with a cutting-edge hackathon challenge, giving you hands-on experience with NVIDIA BioNeMo's tools for small-molecule drug discovery.

### What is BioNeMo?

NVIDIA BioNeMo is a computational drug discovery platform built for developers and data scientists. It provides:
- An open-source machine learning framework for building and training deep learning models
- Optimized and easy-to-use NIM microservices for deploying AI workflows at scale
- Portable deployment through containerized solutions with everything needed for enterprise-grade AI applications

### What is MolMIM?

MolMIM is a generative model for small-molecule drug discovery. It's a probabilistic auto-encoder trained with Mutual Information Machine (MIM) learning, designed to learn an informative and clustered latent space. This allows you to:
- Generate novel molecular structures
- Navigate chemical space intelligently
- Optimize molecules using targeted scoring oracles
- Explore dense latent-space representations of chemical compounds

### What are CDK4 Inhibitors?

CDK4 (Cyclin-Dependent Kinases 4 and 6) are enzymes that play crucial roles in cell division. Inhibiting these proteins has proven to be an effective strategy for treating certain types of cancer, particularly hormone receptor-positive breast cancer.

## 💡 Tips for Success

1. **Follow the Order**: The tutorials build on each other. Complete them sequentially for the best learning experience.

2. **Experiment**: Don't just run the code—modify parameters, try different approaches, and see what happens!

3. **Understand Before Optimizing**: Make sure you understand what each oracle measures before combining them.

4. **Balance is Key**: In drug design, you rarely maximize just one property. Learn to balance multiple objectives.

5. **Document Your Work**: Keep notes on what works and what doesn't. This will help you in the challenge.

6. **Ask Questions**: If something isn't clear, don't hesitate to explore the documentation or ask for help.

7. **Think Like a Drug Designer**: Consider:
   - Can this molecule be synthesized?
   - Will it be safe?
   - Can it reach its target?
   - Will it bind effectively?
